```python
import scipy.special
    
```

```python
def get_binomial_coefficients(n,r,p):
    return scipy.special.comb(n,r) * (p**r) * ((1-p)**(n-r))
```

```python
# Probability of Heads = p
p=2/3
# Total coin tosses = n
n=100

# Resulting probability is sum of coefficients of binomial distribution
probability = 0
for i in range(51):
    probability += get_binomial_coefficients(100,i,p)
    
print("The probability of getting at most 50 heads for 100 coin tosses of biased coin is: {}".format(probability))
```
