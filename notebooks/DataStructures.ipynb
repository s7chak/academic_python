```python
def swap(a,x,y):
    t=a[x]
    a[x]=a[y]
    a[y]=t

def is_sorted(l,lo,hi):
    if lo==hi:
        return True
    elif hi>lo:
        return all(l[i] <= l[i+1] for i in range(lo,hi))
    return False

def getarray(a,lo,hi):
    x=[a[y] for y in range(lo,hi+1)]
    print(x,lo,hi)
```

## Merge Sort

Between 1/2NlogN and NlogN compares

```python
# Merge Sort
def mergeSorted(a,lo,mid,hi):
    aux=[i for i in a]
    i=lo
    j=mid+1
    k=lo
    
    while k<len(a):
        if aux[i]<=aux[j]:
            a[k]=aux[i]
            i+=1
        else:
            a[k]=aux[j]
            j+=1
        k+=1
    return a

def oneMergeSort(a,lo,hi):
    j=int(lo+(hi-lo)/2)
    b=sorted(a[0:j])
    l=sorted(a[j:len(a)])
    a=b+l
    a=mergeSorted(a,0,j,len(a)-1)
    return a
```

## Quick Sort

```python
# Quick Sort
def partition(a):
    if len(a)!=1:
        k=a[0]
        j=len(a)-1
        i=1
        while j>i:
            if a[j]<k and a[i]>=k:
                swap(a,i,j)
            elif a[j]>=k:
                j-=1
            elif a[i]<=k:
                i+=1
        swap(a,j,0)
        return j
    return 0
        
def oneQuickSort(a):
    random.shuffle(a)
    j=partition(a)
    b=sorted(a[0:j])
    l=sorted(a[j:len(a)])
    a=b+l
    return a
```

```python
# Recursive Quick Sort
def recurPartition(a, lo, hi):
    i=lo
    j=hi
    i+=1
    while True:
        while a[i]<a[lo]:
            i+=1
            if i==hi:
                break
        while a[lo]<=a[j]:
            j-=1
            if j==lo:
                break
        if i>=j:
            break
        swap(a,i,j)
    
    swap(a,lo,j)
    return j

def recurQuickSort(a):
    random.shuffle(a)
    recursiveQuickSort(a,0,len(a)-1)
    
def recursiveQuickSort(a, lo, hi):
    if hi<=lo:
        return a
    j=recurPartition(a,lo,hi)
    recursiveQuickSort(a,lo,j-1)
    recursiveQuickSort(a,j+1,hi)
```

```python
def kselect(a,k):
    random.shuffle(a)
    hi=len(a)-1
    lo=0
    while hi>lo:
        j=recurPartition(a,lo,hi)
        if j<k:
            lo=j+1
        elif j>k:
            hi=j-1
        else:
            return a[k]
    return a[k]
        
```

## Quick Sort for Duplicate keys

```python
# Recursive Quick Sort
def recurDupQuickSort(a, shuffle):
    if shuffle:
        random.shuffle(a)
    recursiveDupQuickSort(a,0,len(a)-1)
    
def recursiveDupQuickSort(a, lo, hi):
    
    if hi<=lo:
        return a

    i=lo    
    lt=lo
    gt=hi
    v=a[lo]
    while i<=gt:
        if a[i]<v:
            swap(a,i,lt)
            lt+=1
            i+=1
        elif a[i]>v:
            swap(a,i,gt)
            gt-=1
        else:
            i+=1
    
    recursiveDupQuickSort(a,lo,lt-1)
    recursiveDupQuickSort(a,gt+1,hi)
```

# Binary Search Tree

```python
class Node(object):
    def __init__(self, key, val=0, right=None, left=None):
#         super(Node,self).__init__()
        self.key   = key
        self.val   = val
        self.right = right
        self.left = left

    def printNode(self):
        print(self.key,self.val)
   
    

    
class BSTree(object):
    def __init__(self):
        self.root = None

    def insert(self, key, val=0):
        if self.root is None:
            self.root=Node(key,val)
        else:
            self.root=self._insert(self.root, key, val)
    
    def _insert(self, x, key, val=0):
        if x is not None:
            if key<x.key:
                x.left=self._insert(x.left, key, val)
            elif key>x.key:
                x.right=self._insert(x.right, key, val)
            else:
                x.val=val
            return x
        
        else:
            return Node(key,val)
        
    def height(self):
        if self.root is not None:
            return self._height(self.root, 0)
        else:
            return 0
    
    def _height(self, x, height):
        if x==None:
            return height
        lheight=self._height(x.left, height+1)
        rheight=self._height(x.right, height+1)
        return max(lheight, rheight)
        
        
    def search(self, key):
        if self.root is not None:
            return self._search(self.root, key)
        return False

    def _search(self, x, key):
        if key==x.key:
            return True
        elif key<x.key and x.left!=None:
            return self._search(x.left, key)
        elif key>x.key and x.right!=None:
            return self._search(x.right, key)
        else:
            return False
    
    def minimum(self):
        return self._minimum(self.root)
    
    def _minimum(self, x):
        if x!=None:
            if x.left!=None:
                return self._minimum(x.left)
            else:
                return x

    
    def deletemin(self):
        self.root = self._deletemin(self.root)
        
    def _deletemin(self, x):
        if x.left==None:
            return x.right
        x.left=self._deletemin(x.left)
        return x

    
    
    def delete(self, key):
        if self.root is not None:
            self.root = self._delete(self.root, key)
    
    def _delete(self, x, key):
        if x.key==key:
            if x.left==None:
                return x.right
            elif x.right==None:
                return x.left
            else:
                t=x
                x=self._minimum(t.right)
                x.right=self._deletemin(t.right)
                x.left=t.left
                return x
            
        elif key<x.key and x.left!=None:
            x.left=self._delete(x.left, key)

        elif key>x.key and x.right!=None:
            x.right=self._delete(x.right, key)
        return x
    
    def printTree(self):
        self._printTree(self.root)
    
    
    def _printTree(self,x):
        if x is None:
            pass
        else:
            self._printTree(x.left)
            print(x.key)
            self._printTree(x.right)
        
```

# Sized BST

```python
class node(object):
    def __init__(self, key, val=0, right=None, left=None):
        self.key   = key
        self.val   = val
        self.right = right
        self.left = left
        self.count = 1

    def printNode(self):
        print(self.key,self.val)
   
    

    
class SizedBSTree(object):
    def __init__(self):
        self.root = None
        self.rank = 0

    def size(self):
        return self._size(self.root)

    def _size(self, x):
        if x==None:
            return 0
        else:
            return x.count
        
    def insert(self, key, val=0):
        if self.root is None:
            self.root=node(key,val)
        else:
            self.root=self._insert(self.root, key, val)
    
    def _insert(self, x, key, val=0):
        if x is not None:
            if key<x.key:
                x.left=self._insert(x.left, key, val)
            elif key>x.key:
                x.right=self._insert(x.right, key, val)
            else:
                x.val=val
            x.count=1+self._size(x.left)+self._size(x.right)
            return x
            
        else:
            return node(key,val)
    
    def klargest(self, k):
        c=0
        return self._klargest(self.root, k, c)

    def _klargest(self, x, k, c):

        if x.right == None:
            c+=1
            if c==k:
                return x
            else:
                return self._klargest(x.left, k, c)
        
        else:
            succ=x.right
            while (succ.left != None and succ.left != x):  
                succ = succ.left  

            if (succ.left == None):
                succ.left = x
                return self._klargest(x.right, k, c)

            else:
                succ.left = None
                c += 1
                if c == k:
                    return x
                return self._klargest(x.left, k, c)

    def myKlargest(self, k):    
        self.rank=0
        self._myklargest(self.root, k)

    def _myklargest(self, x, k):
        if x is None:
            return
        else:
            self._myklargest(x.right, k)
            self.rank+=1
            if self.rank==k:
                print(x.key)
            self._myklargest(x.left, k)
            

    def printTree(self):
        self._printTree(self.root)
    
    
    def _printTree(self,x):
        if x is None:
            pass
        else:
            self._printTree(x.left)
            print(x.key)
            self._printTree(x.right)
        
```
