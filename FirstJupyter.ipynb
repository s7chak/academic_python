````python
def extendList(val, list=[]):
    list.append(val)
    return list

list1 = extendList(10)
list2 = extendList(123,[])
list3 = extendList('a')
list4 = extendList('b')

# Weirdness - same list reused
print("list1 = %s" % list1)
print("list2 = %s" % list2)
print("list3 = %s" % list3)
print("list4 = %s" % list4)
````

````python
def extendList2(val, list=None):
    if list==None:
        list=[]
    list.append(val)
    return list

list1 = extendList2(10)
list2 = extendList2(123,[])
list3 = extendList2('a')
list4 = extendList2('b')

# Weirdness - same list reused
print("list1 = %s" % list1)
print("list2 = %s" % list2)
print("list3 = %s" % list3)
print("list4 = %s" % list4)
````

````python
import math
def checkPrime(out):
    rem=-1
    if out==1:
        return False
    if out==2:
        return True
    if out%2==0:
        return False
    
    if out<1000000:
        for div in range(3,int(math.sqrt(out))+1,2):
            if out%div==0:
                return False
    else:
        x=(1<<(out-1))%out
        if x!=1:
            return False
    return True
````
